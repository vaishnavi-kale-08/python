```python
# import libraries
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
# Read the 50 startups.csv
df = pd.read_csv("50_Startups (4).csv")
df.head(4)
```

# 1. Pairplot

```python
# check the pairplot of 50 startups
sns.pairplot(df)
plt.show()
```

# 2. Correlation coefficient

```python
# check the correlation coefficient of 50 startups
df1.corr()
```

# 3.Correlation matrix

```python
sns.heatmap(df1.corr()) 
```
